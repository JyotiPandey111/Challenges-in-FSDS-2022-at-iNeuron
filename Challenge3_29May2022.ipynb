# q9 : Write a code to send a mail to your friend 

```python
# Python code to illustrate Sending mail from
# your Gmail account
import smtplib

# creates SMTP session
s = smtplib.SMTP('smtp.gmail.com', 587)

# start TLS for security
s.starttls()

# Authentication
s.login("sender_email_id", "sender_email_id_password")

# message to be sent
message = "Message_you_need_to_send"

# sending the mail
s.sendmail("sender_email_id", "receiver_email_id", message)

# terminating the session
s.quit()
```

```python
# Python code to illustrate Sending mail
# to multiple users
# from your Gmail account
import smtplib

# list of email_id to send the mail
li = ["xxxxx@gmail.com", "yyyyy@gmail.com"]

for dest in li:
	s = smtplib.SMTP('smtp.gmail.com', 587)
	s.starttls()
	s.login("sender_email_id", "sender_email_id_password")
	message = "Message_you_need_to_send"
	s.sendmail("sender_email_id", dest, message)
	s.quit()
```

# q16 : Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level nesting it should work . 

```python
def dict_to_list(a):
    l = []
    for k,v in a.items():
        l.append(k)
        l.append(v)
    print(l)
        
```

# q20 : write a function by which you will be able to append two PDF files . 

You can use PyPdf2s PdfMerger class. You can simply concatenate files by using the append method. You can pass file handles instead file paths if you want.

```python
from PyPDF2 import PdfMerger

pdfs = ['file1.pdf', 'file2.pdf', 'file3.pdf', 'file4.pdf']

merger = PdfMerger()

for pdf in pdfs:
    merger.append(pdf)

merger.write("result.pdf")
merger.close()
```
